Data Source

```python
lstm_wiki = './LSTM_output/wikitext-2.tsv'
tanh_wiki = './TANH_output/wikitext-2.tsv'
relu_wiki = './RELU_output/wikitext-2.tsv'
gru_wiki = './GRU_output/wikitext-2.tsv'
```

```python
lstm_maptask = './LSTM_output/maptask_transcripts_output'
tanh_maptask = './TANH_output/maptask_transcripts_output'
relu_maptask = './RELU_output/maptask_transcripts_output'
gru_maptask = './GRU_output/maptask_transcripts_output'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

### LSTM OUTPUT

Prepare wikitext data for plotting(LSTM)

```python
lstm_data1 = pd.read_csv(lstm_wiki, sep='\t')
#lstm_data1
```

Prepare map_transcripts data for plotting.Append all output files from maptask_transcripts folder (LSTM)

```python
input_directory = os.fsencode(lstm_maptask)
lstm_data2 = pd.DataFrame()
for file in os.listdir(input_directory):
          filename = os.fsdecode(file)
          if filename.endswith(".tsv"):
                input_path = os.path.join(lstm_maptask, filename)
                df=pd.read_csv(input_path,sep='\t')
                lstm_data2 = lstm_data2.append(df,ignore_index=True)               
lstm_data2
```

Probability distribution curve for wikitext test data (LSTM)

```python

plt.figure(figsize=(15,6),dpi=100)

d_ax = sns.distplot(lstm_data1.PROBABILITY,hist=False,kde_kws={"color": "dodgerblue", "lw": 3})
d_ax.set_title("PROBABILITY DISTRIBUTION CURVE ( wikitext - LSTM )",size = 15)
d_ax.set_xlabel("PROBABILITY",size = 14)
d_ax.set_ylabel("DENSITY",size = 14)

plt.xticks(np.arange(0, 1.01, step=0.04),)
sns.despine(offset=2)
```

Confidence Interval plot with probabilities from wikitest data (LSTM) 

```python
plt.figure(figsize = (10,6), dpi = 100)
sns.set(style = "whitegrid")

# plot using line plot
# lineplot plots the mean of probability for each word position with confidence interval 
mean_axes = sns.lineplot(x = "WORD", y = "PROBABILITY", data = lstm_data1, color='b')

mean_axes.set_title("WORD POSITONS AND PROBABILITY ( wikitext - LSTM ) ")
mean_axes.set_xlabel("word position")
mean_axes.set_ylabel("probability")
plt.xticks(np.arange(0, 502, step=20))
sns.despine(offset=2)
```

Probability distribution curve for maptask_transcripts data (LSTM)

```python
plt.figure(figsize=(15,6),dpi=100)

d_ax = sns.distplot(lstm_data2.PROBABILITY,hist=False,kde_kws={"color": "dodgerblue", "lw": 3})
d_ax.set_title("PROBABILITY DISTRIBUTION CURVE ( maptask - LSTM )",size = 15)
d_ax.set_xlabel("PROBABILITY",size = 14)
d_ax.set_ylabel("DENSITY",size = 14)

#plt.xticks(np.arange(0, 1.01, step=0.04),)
sns.despine(offset=2)
```

Confidence Interval plot with probabilities from maptask_transcripts data (LSTM) 

```python
plt.figure(figsize = (10,6), dpi = 100)
sns.set(style = "whitegrid")

# plot using line plot
# lineplot plots the mean of probability for each word position with confidence interval 
mean_axes = sns.lineplot(x = "WORD", y = "PROBABILITY", data = lstm_data2, color='b')

mean_axes.set_title("WORD POSITONS AND PROBABILITY ( maptask - LSTM ) ")
mean_axes.set_xlabel("word position")
mean_axes.set_ylabel("probability")
#plt.xticks(np.arange(0, 502, step=20))
sns.despine(offset=2)
```
