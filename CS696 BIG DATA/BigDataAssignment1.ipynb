# Task 1

A function, sum_multiples_3_5, that returns the sum of the multiples of 3 and 5 less than N.

for example, if we list all the natural numbers below 20 that are multiples of 3 or 5 but not multiples of both of 3 and 5, we get 3, 5, 6, 9, 10, 12, 18. The sum of these multiples is 63. 



```python
def sum_multiples_3_5(N):
    """Returns the sum of multiples of 3 and 5 less than N but not multiple of both"""
    
    # check if N is not a valid input or if there are no multiples less than N
    if not (isinstance(N, int)):
        raise TypeError
    elif N <= 0:
        raise ValueError
    else:
                   
        #sum of multiples of only 3
        added_value = sum([multiple_of_3 for multiple_of_3 in range(0, N, 3) if(multiple_of_3 % 5 != 0)])
        
        #find sum of multiples of only 5 and update previous sum    
        added_value += sum([multiple_of_5 for multiple_of_5 in range(5, N, 5) if(multiple_of_5 % 3 != 0)])
        
        return(added_value)
    
```

# Task 2

A function, pattern_count, returns the number of times the pattern occurs in the text.

Function pattern_count has two arguments. 
The first arguments is a string, lets call it text, and the second argument is also a string, call it pattern. 

For example, pattern_count(“abababa”, “aba”) == 3, 
pattern_count(“aaaaa”, “aa”) == 4, 
pattern_count(“Abcde”, “abc”) == 0

```python
def pattern_count(text, pattern):
    """Returns the number of times the pattern occurs in the text"""
    
    start_index = 0
    pattern_count = 0
    end_index = len(text)
    
    if text and pattern:
        
        if str(text).isalpha():
            
            while True:
                #find index of first occurrence of pattern between start and end
                index = text.find(pattern, start_index, end_index)
                if index != -1 :
                    # increment counter if pattern found
                    pattern_count = pattern_count + 1
                    start_index = index + 1
                else:
                    break
            if pattern_count > 0 :
                return(pattern_count)
            else:
                return False
        else:
            raise TypeError
    
    else:
        raise ValueError
```

# Task 3

A function to find the n'th positive octaldrome. 

A palindrome is a string that reads the same forwards and backwards. 
A number is octaldrome if its octal representation is a palindrome. 
So 945 is an octaldrome as in octal it is 1661. The first nine positive octaldromes are 1, 2, 3, 4, 5, 6, 7, 9 and 18.

For example the 1'st positive octaldrome is 1, the fifth positive octaldrome is 5 and the eighth positive octaldrome is 9.  
So we have
nth_octaldrome(1) ==1 ,
nth_octaldrome(9) == 18.

```python
def nth_octaldrome(n):
    """Returns the nth positive octaldrome"""
    
    # check if the input of required rank is valid
    if (n<1):
        raise ValueError
    elif not isinstance(n, int):
        raise TypeError
    else:
        number = 1
        octaldrome_rank = 0
        
        while True:
            # calculate octal value
            octal_value = oct(number)[2:]
            
            # check if it is octaldrome
            if(octal_value == "".join(reversed(octal_value))):
                octaldrome = number
                octaldrome_rank += 1
            # check if octaldrome rank is the required rank n  
            if octaldrome_rank == n:
                return(octaldrome)
            else:
                number += 1
```

# Task 4
Aa function, most_frequent_word, returns the sequence word in s of length n that occurs most in the string. 

Function has two arguments. The first argument is a string, the second argument is an integer, n. 

For example, most_frequent_word(“TCGAAGCTAGACGCTAGTAGCTAGTGTGCA”,4)    returns (“CTAG”,“GCTA”) 

```python
def most_frequent_word(string, n):
    """Returns the most frequent word in the string of length n"""
    
    max_count=0
    word_list = []
    
    # check if string is not empty and input length is valid
    if string:
        if str(string).isalpha():
            
            if n > 0 and isinstance(n, int):
                  
                for index in range(len(string)-n+1):
                    substring = string[index : index + n]
                    sub_count = sum(substring == string[sub_index : sub_index + n] for sub_index in range(index-n+1))
            
                    #if new substring has same number of occurences as the previous substring
                    if sub_count == max_count:
                        if substring not in word_list:
                            word_list.append(substring)
                            
                    #if the new substring has more occurences than the previous one   
                    elif sub_count > max_count :
                        word_list.clear()
                        if substring not in word_list:
                            word_list.append(substring)
                        #update max_count
                        max_count = sub_count
                return(word_list)
            else:
                raise ValueError
        else:
            raise TypeError
    else:
        raise ValueError
```

# Task 5
A function reverse_complement(s) that returns the reverse complement of a DNA string s.

In DNA strings, symbols 'A' and 'T' are complements of each other, as are 'C' and 'G'. 
The reverse complement of a DNA string s is the string formed by reversing s, then taking the complement of each symbol.
For example, the reverse complement of "GTCA" is "TGAC"


```python
def reverse_complement(s):
    """Returns the reverse complement of a DNA string"""
    
    rev_complement = []
    complement_dict = {'A' : 'T', 'C' : 'G', 'T' : 'A', 'G' : 'C'}
    
    #Check if s is a valid DNA string or an empty string
    if not s or any(i not in ['A','C','T','G'] for i in s):
        raise ValueError
    elif str(s).isdigit():
        raise TypeError
    else:        
        #find reverse of string s
        reversed_string = "".join(reversed(s))
        
        #find complement of the reversed string
        for i in reversed_string:
            rev_complement.append(complement_dict.get(i))
        return("".join(rev_complement))
```

# Task 6
A function gc_content(s) with input a DNA-string and returns the GC-content of the string.

(The GC-content of a DNA string is given by the percentage of characters in the string that are 'C' or 'G'. 
For example, the GC-content of "AGCTATAG" is 37.5%. So, gc_content("AGCTATAG") == 0.375 )

```python
def gc_content(s):
    """Returns the GC content of a DNA string"""
    
    #check if string is empty or if it is not a valid DNA string
    if not s or any(i not in ['A','C','T','G'] for i in s): 
        raise ValueError
    elif str(s).isdigit():
        raise TypeError
    
    else:
        #calculate GC content
        gc_value = (s.count('G') + s.count('C')) / len(s)
        return(gc_value)
```

# Task 7

A function that receives a collection of strings and a map from strings to integers. 
Returns a collection of integers that are values of the map corresponding to one of the strings in the collection. 

For example, given Array("Tom", "Fred", "Harry") and Map("Tom" -> 3,"Dick" -> 4, "Harry" -> 5), returns Array(3, 5). 



```python
def key_values(array, map_dictionary):
    """Returns values of the map corresponding to the strings in the array"""
    
    array_string_values = []
    flag = 0
    
    # check if inputs are non-empty
    if array and map_dictionary:
        # check if array contains strings
        if all(str(string).isalpha() for string in array):
            # check if dictionary keys are all strings
            if all(str(key).isalpha() for key in map_dictionary.keys()):
                # check if dictionary values are integers
                if all(isinstance(values,int) for values in map_dictionary.values()):
                    
                    for i in array:
                        #fetch value of the key after checking if the key exists
                        if i in map_dictionary:
                            array_string_values.append(map_dictionary.get(i))
                        else:
                            # To know if there are no strings in array that match the map
                            flag = flag + 1
                    if flag == len(array) :
                        return False
                    else:
                        return(array_string_values)
                else:
                    raise TypeError
            else:
                raise ValueError
        else:
            raise TypeError
    else:
        raise ValueError
            
  
```
